```python
import pandas as pd
import nltk
import plotly.express as px
import matplotlib.pyplot as plt

from transformers import pipeline
from pysentimiento import create_analyzer
from nltk.corpus import stopwords
from wordcloud import WordCloud
```

```python
classificador_sentimento = pipeline(
    "sentiment-analysis",
    model="distilbert-base-uncased-finetuned-sst-2-english"
)

modelo_analise_sentimento = create_analyzer(
    task="sentiment",
    lang="pt"
)
```

```python
texto_teste = """
A fritadeira é sensacional, muito além do que imaginei.
Fácil de limpar e muito eficiente.
"""

resultado = modelo_analise_sentimento.predict(texto_teste)

print(resultado)
```

```python
dados = pd.read_csv('https://raw.githubusercontent.com/alura-cursos/hugging_face/refs/heads/main/Dados/resenhas.csv')

dados.head()
```

```python
resultados_previsao = modelo_analise_sentimento.predict(dados['Resenha'])

dados["Sentimento"] = [
    resultado.output
    for resultado in resultados_previsao
]
```

```python
df_sentimento = (
    dados.groupby("Sentimento")
    .size()
    .reset_index(name="Contagem")
)

fig = px.bar(
    df_sentimento,
    x="Sentimento",
    y="Contagem",
    title="Distribuição dos Sentimentos"
)

fig.show()
```

```python
nltk.download("stopwords")

portuguese_stopwords = set(
    stopwords.words("portuguese")
)
```

```python
def gerar_nuvem_palavras(
    dataframe,
    coluna_texto,
    sentimento
):

    textos = dataframe.query(
        f"Sentimento == '{sentimento}'"
    )[coluna_texto]

    texto_unido = " ".join(textos)

    palavras = texto_unido.split()

    palavras_filtradas = [
        palavra
        for palavra in palavras
        if palavra.lower() not in portuguese_stopwords
    ]

    texto_filtrado = " ".join(palavras_filtradas)

    nuvem = WordCloud(
        width=800,
        height=500,
        max_words=50
    ).generate(texto_filtrado)

    plt.figure(figsize=(10, 7))
    plt.imshow(nuvem)
    plt.axis("off")
    plt.show()
```

```python
gerar_nuvem_palavras(
    dados,
    "Resenha",
    "POS"
)

gerar_nuvem_palavras(
    dados,
    "Resenha",
    "NEG"
)
```
